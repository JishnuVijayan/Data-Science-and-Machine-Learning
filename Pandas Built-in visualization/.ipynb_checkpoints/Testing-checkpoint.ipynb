### Style sheets

Matplotlib has [style sheets](http://matplotlib.org/gallery.html#style_sheets) you can use to make our plots look a little nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They basically create a set of style rules that your plots follow. They make all our plots have the same look and feel more professional.

# Plot types

There are several plot types built-in to pandas.
* `df.plot.area()`
* `df.plot.barh()`
* `df.plot.density()`
* `df.plot.hist()`
* `df.plot.line()`
* `df.plot.scatter()`
* `df.plot.bar()`
* `df.plot.box()`
* `df.plot.hexbin()`
* `df.plot.kde()`
* `df.plot.pie()`


We can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)

### Area

```python
df2.plot.area(alpha = 0.4)
```

### Bar

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)
```

### Histogram

```python
df1['A'].plot.hist()
```

```python
df1['A'].plot.hist(bins = 30)
```

### Line plot

```python
df1.plot.line( y = 'B', figsize=(12, 3))
```

```python
df1.plot.line( y = 'B', figsize=(12, 3), lw = 1)
```

### Scatter plot

```python
df1.plot.scatter(x = 'A', y = 'B')
```

We can use c to color based off another column value

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['C'])
```

```python
df1.plot.scatter(x = 'A', y = 'B', s = df1['C']*200)
```

### Box

```python
df2.plot.box()
```

### Hexagonal bin

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns = ['a', 'b'])
df.plot.hexbin(x = 'a', y = 'b')
```

```python
df.plot.hexbin(x = 'a', y = 'b', cmap = 'Oranges', gridsize = 25)
```

### Kernal density estimation plot

```python
df2['a'].plot.kde()
```

### density

```python
df2.plot.density()
```
