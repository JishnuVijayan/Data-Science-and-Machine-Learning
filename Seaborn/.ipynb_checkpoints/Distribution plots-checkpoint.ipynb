# Distribution plots

Distribution plots allows us to visualize the distribution of a data set. 
* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# displot

`displot()` shows the distribution of a **univariate** set of observations

```python
sns.displot(tips['total_bill'])
```

From the above graph we can understand that most of the total bills are in between 10 to 20.

```python
sns.displot(tips['total_bill'], bins = 40, kde = True)
```

kde is kernel density estimation which shows the densit(highest density is between 10 to 20)

# jointplot

`jointplot()` allows you to basically match up two displots for bivariate data with our choice of what `kind` of parameter to compare with:

* scatter
* reg
* resid
* kde
* hex

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
```

```python
 sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg') # 'reg' means regression
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'resid')
```

From all this graph we can understand that tip is directly propotional to total bill.

# pairplot

`pairplot()` will plot pairwise relationships across the entire dataframe(for numerical columns) and supports a color `hue` argument (for categorical columns).

```python
sns.pairplot(tips)
```

The first plot is 'total_bill' vs 'total_bill' that is why it is a histogram. Like this there are 2 more in this pair plot

```python
sns.pairplot(tips, hue = 'sex')
```

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

# rugplot

`rugplot` just draw a dash mark for every point on a univariate distribution. They are building block of `kdeplot()`

```python
sns.rugplot(tips['total_bill'])
```

# kdeplot

`kdeplot()` are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These kde plots replace every single observation with a Guassian(Normal) distribution around the value. 

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
