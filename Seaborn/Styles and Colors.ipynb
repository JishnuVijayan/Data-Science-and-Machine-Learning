```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Styles

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_style('white')
sns.countplot(x = 'sex', data = tips)
```

## spine removal

```python
sns.countplot(x = 'sex', data = tips)
sns.despine(left = True)
```

```python
sns.countplot(x = 'sex', data = tips)
sns.despine(left = True, bottom = True)
```

## Size and Aspect

we can use matplotlib's **plt.figure(figsize=(width,height)** to change the size of most seaborn plots.

we can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
plt.figure(figsize=(12, 3))
sns.countplot(x = 'sex', data = tips)
```

```python
sns.lmplot(x='total_bill',y='tip',aspect=4,data=tips)
```

## scale and context

The `set_context()` allows us to override default parameters

```python
sns.set_context('poster', font_scale = 1)
sns.countplot(x = 'sex', data = tips, hue = 'sex', palette = 'Set3')
```
