# cufflinks and iplot()

We can zoom, select, deselect, download etc... In short we can say that this is interactive plots

Using cufflinks and `iplot()` we can make:
* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

### `scatter`

```python
df.iplot(kind = 'scatter')
```

```python
df.iplot(kind = 'scatter', x = 'A', y = 'B', mode = 'markers', size = 10)
```

### `bar`

```python
df2.iplot(kind = 'bar', x = 'category', y = 'values')
```

```python
df.count().iplot(kind = 'bar')
```

```python
df.sum().iplot(kind = 'bar')
```

### `box`

```python
df.iplot(kind = 'box')
```

### 3-d surface

```python
df3 = pd.DataFrame({'x':[1, 2, 3, 4, 5], 'y': [10, 20, 30, 40, 50], 'z': [5, 4, 3, 2, 1]})
```

```python
df3.iplot(kind = 'surface', colorscale = 'rdylbu')
```

### `hist`

```python
df['A'].iplot(kind = 'hist')
```

```python
df['A'].iplot(kind = 'hist', bins = 50)
```

### `bubble`

```python
df.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C')
```

### `scatter_matrix`

```python
df.scatter_matrix()
```
